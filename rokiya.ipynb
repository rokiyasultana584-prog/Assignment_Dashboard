**H1: Sum of numbers from 1 to** **N**

```
def sum_to_n(n):
    return sum(range(1, n + 1))

print(sum_to_n(10))
```

**H2: Square of a number**

```
def square(n):
    return n ** 2

print(square(5))
```

**H3: Check if a number is even or odd**

```
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(4))
print(even_or_odd(7))
```
