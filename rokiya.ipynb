**H1: Sum of numbers from 1 to** **N**

```
def sum_to_n(n):
    return sum(range(1, n + 1))

print(sum_to_n(10))
```

**H2: Square of a number**

```
def square(n):
    return n ** 2

print(square(5))
```

**H3: Check if a number is even or odd**

```
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(4))
print(even_or_odd(7))
```

**H4: Return the larger of two numbers **

```
def larger(a, b):
    if a > b:
        return a
    else:
        return b

print(larger(8, 3))
print(larger(5, 10))
```

**H4: Return the larger of two numbers**

```
def larger(a, b):
    return max(a, b)

print(larger(8, 3))
print(larger(5, 10))
```

**H6: Find all unique elements in a list**

```
li = [1, 2, 2, 3, 4, 4, 5]
unique = list(set(li))
print(unique)
```

**H9: Find the key with the maximum value in a dictionary**

```
d = {'a': 10, 'b': 20, 'c': 15}
max_key = max(d, key=lambda k: d[k])
print(max_key)
```
